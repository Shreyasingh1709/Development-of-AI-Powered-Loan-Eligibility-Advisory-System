```python
for col in df.columns:
    print(f"{col:25} | dtype: {str(df[col].dtypes):10} | unique: {df[col].nunique():5} | missing: {df[col].isna().sum():5}")
```

```python
# Columns with mixed types or stored as object but look numeric
for col in df.select_dtypes(include='object'):
    try:
        df[col].astype(float)
        print(f"⚠️ Column '{col}' seems numeric but stored as object")
    except:
        pass
```

```python
numeric_col = df.select_dtypes(include = ['int64','float64']).columns
print("Numeric columns: ",list[numeric_col])
```

```python
for col in numeric_col:
    unique_values = df[col].unique()
    invalid_values = [v for v in unique_values if isinstance(v, str)]
    if invalid_values:
        print(f"⚠️ Column '{col}' has non-numeric values: {invalid_values[:10]}")
    else:
        print(f"✅ Column '{col}' is clean (all numeric)")
```
