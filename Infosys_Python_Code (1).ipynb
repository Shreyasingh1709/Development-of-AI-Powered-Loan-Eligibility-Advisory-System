```python
# Columns with mixed types or stored as object but look numeric
for col in df.select_dtypes(include='object'):
    try:
        df[col].astype(float)
        print(f"âš ï¸ Column '{col}' seems numeric but stored as object")
    except:
        pass
```

```python
numeric_col = df.select_dtypes(include = ['int64','float64']).columns
print("Numeric columns: ",list[numeric_col])
```

```python
for col in numeric_col:
    unique_values = df[col].unique()
    invalid_values = [v for v in unique_values if isinstance(v, str)]
    if invalid_values:
        print(f"âš ï¸ Column '{col}' has non-numeric values: {invalid_values[:10]}")
    else:
        print(f"âœ… Column '{col}' is clean (all numeric)")
```

```python
# Check again after cleaning
clean_check = {col: {type(v).__name__ for v in df[col].dropna().unique()}
               for col in df.columns}
clean_check = {col: t for col, t in clean_check.items() if len(t) > 1}

if clean_check:
    print("\nâš ï¸ Still mixed types remain in:", clean_check)
else:
    print("\nâœ… All columns are now clean and consistent!")
```

```python
df.to_csv("cleaned_dataset.csv", index=False)
print("\nðŸ’¾ Cleaned dataset saved as 'cleaned_dataset.csv'")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram for all numeric columns
df.hist(figsize=(10, 8))
plt.show()

# Correlation heatmap
# Exclude non-numeric columns before calculating correlation
numeric_df = df.select_dtypes(include=['int64', 'float64'])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.show()

# Pairplot for feature relationships
# Select a subset of numeric columns for pairplot to avoid overcrowding
pairplot_cols = ['Age', 'Monthly_Income', 'Loan_Amount_Requested', 'Credit_Score', 'Loan_Approved']
sns.pairplot(df[pairplot_cols])
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Correct the column name from 'Loan_Amount' to 'Loan_Amount_Requested'
X = df[['Monthly_Income', 'Loan_Amount_Requested']]  # example features
y = df['Loan_Approved']                    # target variable

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = LinearRegression()
model.fit(X_train, y_train)

print("RÂ² Score:", model.score(X_test, y_test))
```
